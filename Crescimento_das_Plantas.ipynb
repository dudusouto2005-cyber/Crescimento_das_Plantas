# Avaliação do Crescimento das Plantas

## Informações do experimento

**Cultura:** Tomate (*Solanum lycopersicum*)

**Material genético:** San Marzano

**Área experimental:** 42 m²

**Número de plantas avaliadas:** 5

## Objetivo

Avaliar o crescimento inicial das plantas por meio da medição da altura, verificando a uniformidade do desenvolvimento da cultura.

```python
# ======================================
# CRESCIMENTO DAS PLANTAS
# ======================================

alturas = [42, 45, 41, 47, 44]

media = sum(alturas) / len(alturas)

print("RELATÓRIO DE CRESCIMENTO")
print("-" * 40)

for planta, altura in enumerate(alturas, start=1):
    print(f"Planta {planta}: {altura} cm")

print("-" * 40)

print(f"Altura média: {media:.1f} cm")
print(f"Maior altura: {max(alturas)} cm")
print(f"Menor altura: {min(alturas)} cm")

if media >= 44:
    print("\nDesenvolvimento satisfatório.")
else:
    print("\nDesenvolvimento abaixo do esperado.")
```

# Interpretação

Foram avaliadas cinco plantas representativas da área experimental de 42 m².

A altura média observada foi de aproximadamente 44 cm, indicando crescimento uniforme entre as plantas.

Essa uniformidade demonstra que o manejo adotado durante o cultivo, incluindo irrigação e adubação, proporcionou condições favoráveis ao desenvolvimento inicial da cultura.

O acompanhamento da altura das plantas permite identificar precocemente possíveis problemas de manejo e auxilia na tomada de decisões durante o ciclo produtivo.
